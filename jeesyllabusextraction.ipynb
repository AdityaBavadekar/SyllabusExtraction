This is script that extracts jee mains syllabus from PDF file from Jee Mains Website and creates a file for it.

```python
url = "https://jeemain.nta.nic.in/information-bulletin/"
```

```python
FILE = "JEE Mains Syllabus"
```

```python
import requests
from bs4 import BeautifulSoup

response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")
links = []

for link in soup.find_all("a"):
    href = link.get("href")
    if href:
        links.append(href)

pdfs = []
print("Links on the page:")
for link in links:
    if link.__contains__("pdf"):
        pdfs.append(link)
        print(link)
selection = sorted(pdfs)[-1]
print(selection)
```

```python
btdata = requests.get(selection)

with open(FILE+".pdf", "wb") as f:
    f.write(btdata.content)

print("PDF file downloaded and saved to disk.")
```

```python
import PyPDF2

with open(FILE+".pdf", "rb") as f:
    pdf_reader = PyPDF2.PdfReader(f)
    num_pages = len(pdf_reader.pages)
    i = 0
    i2 = 0
    for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]
        page_text = page.extract_text()

        print(f"Page {page_num + 1}:")
        print(len(page_text))
        if page_text.__contains__("SYLLABUS for JEE (Main)"):
             i = page_num
             print(page_text)
        if page_text.__contains__("Syllabus for Paper 2A"):
             i2 = page_num
             
    with open(FILE+".txt", "w") as fw:
        for pn in range(i,i2):
            page_text = pdf_reader.pages[pn].extract_text()
            fw.write(page_text+" ")
```

```python
with open(FILE+".txt","r") as f:
    lines = f.readlines()
lines
```

```python

syllabus = []
current_sub = ""
for l in lines:
    if l.__contains__("UNIT"):
        l = l.lstrip()
        index = l.find(":")
        unit = l[:index]
        mydict = [current_sub,unit,l[index+1:].replace("\n","")]
        syllabus.append(mydict)
        print(mydict)
    else:
        l=l.replace(" ","").replace("\n","")
        subject = ""
        if l == "MATHEMATICS" or l == "PHYSICS" or l == "CHEMISTRY":
            subject = l
            print(l)
        if l[2:] == "MATHEMATICS" or l[2:] == "PHYSICS" or l[2:] == "CHEMISTRY":
            subject = l[2:]
            print(l[2:])
        if subject!= "":
            current_sub = subject
```

```python

import csv
csv_file = "JEE Mains Syllabus.csv"
with open(FILE+".csv", "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Subject","Unit", "Topic"])
        for i,s in enumerate(syllabus):
            writer.writerow(s)
```
